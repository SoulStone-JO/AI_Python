```python
import tensorflow as tf
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
```

```python
n_epochs = 46500 # 迭代次数
learning_rate = 0.001
```

```python
housing = fetch_california_housing(data_home="C:/Users/Jo rakuten/scikit_learn_data",download_if_missing=True)
m, n = housing.data.shape
print(m, n)
housing_data_plus_bias = np.c_[np.ones((m, 1)), housing.data]
```

对数据进行归一化处理，StandardScaler()提供标准归一化，包括方差归一化和均值归一化。其中，z = (x - u) / s ；

```python
scaler = StandardScaler().fit(housing_data_plus_bias)  
```

```python
print(scaler.mean_)  # StandardScaler()的attribute：mean_，为每列特征的均值         
```

```python
print(scaler.var_)  # StandardScaler()的attribute：var_，为每列特征的方差
```

```python
print(scaler.scale_)
```

```python
print(scaler.transform(housing_data_plus_bias))   # 显示转换的数据
```

```python
scaled_housing_data_plus_bias = scaler.transform(housing_data_plus_bias)
```

```python
X = tf.constant(scaled_housing_data_plus_bias,dtype=tf.float32,name="X")
y = tf.constant(housing.target.reshape(-1, 1),dtype=tf.float32,name="y")
```

random_uniform函数创建图里一个节点包含随机数值，给定它的形状和取值范围，就像numpy里面rand()函数,uniform表示均匀分布。

```python
theta = tf.Variable(tf.random_uniform([n + 1, 1],-1.0,1.0), name="theta")
```

```python
y_pred = tf.matmul(X,theta, name="predictions")
```

```python
error = y_pred - y # 误差公式 ERROR = y(hat) - y
```

```python
rmse = tf.sqrt(tf.reduce_mean(tf.square(error), name="rmse")) # 误差的平方和之后再求均值
```

梯度的公式：(y_pred - y) * xj

```python
gradients = 2/m * tf.matmul(tf.transpose(X), error)
```

赋值函数对于BGD来说就是 theta_new = theta - (learning_rate * gradients)

```python
training_op = tf.assign(theta, theta - learning_rate * gradients) # tf.assign为赋值，这步进行梯度下降和迭代
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)

    for epoch in range(n_epochs):
        if epoch % 100 == 0:
            print("Epoch", epoch, "RMSE = ", rmse.eval())
        sess.run([training_op])

    best_theta = theta.eval()
    print(best_theta)
    print(rmse)
```

从最后来看，误差在2.191632之间发生了震荡
